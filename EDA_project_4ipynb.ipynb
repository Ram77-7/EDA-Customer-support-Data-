```python
df.groupby('channel_name')['Item_price'].mean()
```

Here we find the mean values for the Email ,inbound , outcall

```python
def itprice(X):
  if X=='Email':
    return 6691.8
  elif X=='Inbound':
    return 5622.4
  else:
    return 5360.5
```

Here I write a function to replace the values based on mean values .

```python
A['Item_price']=A['channel_name'].apply(itprice)
```

Here I used apply() function to apply the above function of mean values to the Item_price based on channel_name.

```python
A
```

```python
C=pd.concat([A,B],ignore_index=True)
```

Here We combined the both tables and permanently makes changes for this C data frame.

```python
C
```

```python
C.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Here import visualization packages are MatplotLib.pypolt and seaborn

```python
sns.barplot(x='channel_name',y='Item_price',data=C)
plt.xlabel('channel_name')
plt.ylabel('Item_price')
plt.title('channel_name vs Item_price')
plt.show()
```

Here we display barplot based on channel_name vs Item_price.

```python
sns.lineplot(x='Agent Shift',y='Item_price',data=C)
plt.xlabel('Agent shift')
plt.ylabel('Item_price')
plt.title('Agent shift vs Item_price')
plt.show()
```

Here we used the lineplot for the item_price vs Agent shift.

```python
sns.boxplot(x='channel_name',y='CSAT Score',data=C)
plt.xlabel('channel_name')
plt.ylabel('CSAT Score')
plt.title('channel_name vs CSAT Score')
plt.show()
```

Here is the visualization box plot represents the CSAT score vs channel_name.
